```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.interpolate import UnivariateSpline
```

```python
### Generate Random Variable Sequence
n = 100000 # Numero de R.V.s
N = 1000 # Numero de muestras en cada R.V.
media,varianza = 36,49 # media y varianza de R.V.s
X_n = np.zeros((n,N))

suma = X_n[0]
for i in np.arange(n):
    X_n[i] = np.random.normal(media,np.sqrt(varianza),size=N)
    suma += X_n[i]
    
suma

Xbar_n = (1/n)*suma

# Definicion de la RV
RV = np.sqrt(n)*(Xbar_n-media)/np.sqrt(varianza)

mu, sigma = 0, 1 # mean and standard deviation
count, bins, ignored = plt.hist(RV, 30, density=True)
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) *
               np.exp( - (bins - mu)**2 / (2 * sigma**2) ),
         linewidth=2, color='r')

plt.show()
```
